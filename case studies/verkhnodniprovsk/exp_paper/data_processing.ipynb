```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats as ss
```

```python
df = pd.read_csv('results_verkhnodniprovsk.txt', delimiter='\t', header=None)
```

```python
plt.hist(df[0])
plt.show()
```

```python
def needed(xs, alpha=0.05):
    mean = xs.mean()
    var = xs.var()
    u_alpha = ss.norm.ppf(1 - alpha)
    return var * u_alpha**2 / (alpha * mean)**2
```

```python
needed(df[0])
```

```python
data = pd.read_csv('verkhnodniprovsk-best.csv', delimiter='\t', header=None)
data
```

```python
schedules = []
for d in range(10):
    schedules.append([data[1][d], data[2][d], data[3][d], data[4][d], data[5][d]])
schedules
```

```python
schedules.append([12, 4, 8, 26, 5])
```

```python
basic = [75, 90, 30, 95, 0]
optimized = [3, 7, 2, 26, 3]
bdf = pd.read_csv(f'./verkhnodniprovsk/{basic}+B.txt', delimiter='\t', header=None)
print(basic, '\t', bdf[0].mean(), bdf[1].mean(), needed(df[0]))
basic_data = bdf[1]

odf = pd.read_csv(f'./verkhnodniprovsk/{optimized}+w.txt', delimiter='\t', header=None)
print(optimized, '\t', odf[0].mean(), odf[1].mean(), needed(df[0]))
optimized_data = odf[1]

print(f'effect={(bdf[0].mean() - odf[0].mean())/bdf[0].mean()}')
```

```python
results = []
for schedule in schedules:
    df = pd.read_csv(f'./verkhnodniprovsk/{schedule}+w.txt', delimiter='\t', header=None)
    print(schedule, '\t', df[0].mean(), df[1].mean(), needed(df[0]), df[3].mean())
    results.append([schedule, df[0].mean(), df[1].mean(), needed(df[0]), df[3].mean()])
```

```python
results.sort(key=lambda x: x[1])
for result in results:
    print(result[0], '\t', round(result[2], 2), '\t', round(result[4], 0))
```

```python
bins, opacity = 8, 0.65

plt.figure(figsize=(8, 5))
plt.hist(basic_data, bins=bins, alpha=opacity, color='blue')
plt.hist(optimized_data, bins=bins, alpha=opacity, color='green')
plt.legend(['Basic schedule', 'Optimized schedule'])

plt.axvline(basic_data.mean(), linestyle='--', color='orange')
plt.axvline(optimized_data.mean(), linestyle='--', color='red')
plt.xlabel('Total waiting time [min.]')
plt.ylabel('Frequency [-]')

plt.show()
```
